```python
import pandas
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pandas.read_csv('train.csv')
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Sobrevivência por Gênero')
plt.show()
```

Mulheres tiveram uma taxa de sobrevivência significativamente maior que os homens. Isso reforça a política 'mulheres e crianças primeiro'

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Sobrevivência por classe')
plt.show()
```

A taxa de sobrevivência foi maior na 1ª classe, sugerindo uma preferência no resgate ou acesso facilitado aos botes salva-vidas.

```python
sns.histplot(data=df, x='Age', hue='Survived', bins=30)
plt.title('Sobrevivência por Idade')
plt.show()
```

Passageiros entre 0 e 10 anos tiveram uma taxa alta de sobrevivência. Idosos sobreviveram menos.

```python
sns.boxplot(data=df, x='Pclass', y='Fare', hue='Survived')
plt.title('Tarifa x Classe X Sobrevivência')
plt.show()
```

Passageiros que pagaram tarifas mais altas — geralmente da 1ª classe — tiveram maior taxa de sobrevivência.

```python
sns.countplot(data=df, x='Embarked', hue='Survived')
plt.title('Sobrevivência por Porto de embarque')
plt.show()
```

Entre os portos, Cherbourg teve maior proporção de sobreviventes — possivelmente por embarcar mais passageiros da 1ª classe.
